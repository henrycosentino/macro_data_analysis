## Defining the Data
- Define **NFP Survey Delta** as the difference between the actual data release and Bloomberg's survey estimate of Wall Streets forecast:
  - A positive NFP Survey Delta means the NFP change was higher than Wall Street forecasted.
    - This can indicate economic strength or strengthening, potentially causing rates to sell off.
  - A negative NFP Survey Delta means the change in NFP was lower than forecasted by Wall Street.
    - This can indicate economic weakness or weakening, potentially causing rates to rally.
  
- Define **UE Survey Delta** as the difference between the actual unemployment data release and Bloomberg's survey estimate:
  - A positive UE Survey Delta means the unemployment rate was higher than forecasted by Wall Street.
    - This can indicate economic weakness or weakening, potentially causing rates to rally.
  - A negative UE Survey Delta means the unemployment rate was lower than forecasted by Wall Street.
    - This can indicate economic strength or strengthening, potentially causing rates to sell off.

- Define **Actual Release Date** as the date the data was actually released and not the period (month) the data is being reported for

- Calculate the **10YR bps Delta** as the daily change in the 10-Year U.S. Treasury yield (USGG10YR Index) on the release date

- Calculate the **2YR bps Delta** as the daily change in the 2-Year U.S. Treasury yield (USGG2YR Index) on the release date

```python
# Transforming into a %

jobs_df['UE Survey Delta (%)'] = jobs_df['UE Survey Delta'] * 100
jobs_df = jobs_df.drop(columns='UE Survey Delta')
jobs_df
```

```python
# Preparing the data for scatter plot

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# 10YR Frame
scaled_nfp_10yr = scaler.fit_transform(jobs_df[['NFP Survey Delta', '10YR bps Delta']])
df_nfp_10yr = pd.DataFrame(scaled_nfp_10yr, columns=jobs_df[['NFP Survey Delta', '10YR bps Delta']].columns)

# 2YR Frame
scaled_nfp_2yr = scaler.fit_transform(jobs_df[['NFP Survey Delta', '2YR bps Delta']])
df_nfp_2yr = pd.DataFrame(scaled_nfp_2yr, columns=jobs_df[['NFP Survey Delta', '2YR bps Delta']].columns)
```

```python
# Outlier Function

def outlier(frame, column, lower, upper):
    filtered_frame = frame[(frame[column] >= lower) & (frame[column] <= upper)]
    return filtered_frame
```

```python
# Grouping the UE Data
ue_df = jobs_df[['Actual Release Date', 'UE Survey Delta (%)', '10YR bps Delta']].copy()

# Positive UE Survey Delta means the unemployment rate was higher than surveyed by Wall Street
# Negative UE Survey Delta means the unemployment rate was lower than surveyed by Wall Street

bins = [-np.inf, -0.4, -0.3, -0.2, -0.1, 0, 0.1, 0.2, 0.3, 0.4, np.inf]
ue_df['UE Survey Delta (%) Binned'] = pd.cut(ue_df['UE Survey Delta (%)'], bins=bins)

ue_grouped = ue_df.groupby('UE Survey Delta (%) Binned', observed=True)
ue_grouped_10yr = ue_grouped['10YR bps Delta']
ue_grouped_10yr.agg(['mean', 'median', 'std', 'count']).round(2)
```

```python
# Removing Outliers for Nonfarm Payrolls

jobs_df['z_score'] = np.abs(stats.zscore(jobs_df['NFP Survey Delta']))

# Drop rows where Z-score exceeds the threshold
threshold = 4
jobs_df_no_outliers = jobs_df[jobs_df['z_score'] <= threshold].drop(columns=['z_score'])

# Calculate Q1, Q3, and IQR
q1 = jobs_df['NFP Survey Delta'].quantile(0.25)
q3 = jobs_df['NFP Survey Delta'].quantile(0.75)
iqr = q3 - q1

# Define the lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Drop rows where 'NFP Survey Delta' is outside these bounds
jobs_df_no_outliers_nfp = jobs_df[(jobs_df['NFP Survey Delta'] >= lower_bound) & 
                              (jobs_df['NFP Survey Delta'] <= upper_bound)]

jobs_df_no_outliers_nfp = jobs_df_no_outliers_nfp.reset_index(drop=True)
jobs_df_no_outliers_nfp.head()
```

```python
# Grouping Nonfarm Payrolls Data

nfp_df = jobs_df_no_outliers_nfp[['Actual Release Date', 'NFP Survey Delta', '10YR bps Delta', '2YR bps Delta']].copy()

# Positive NFP Survey Delta means the Change in NFP was higher than surveyed by Wall Street --> Good sign --> Positive Growth --> Rise in rates
# Negative NFP Survey Delta means the Change in NFP was lower than surveyed by Wall Street --> Bad sign --> Negative Growth --> Decline in rates

bins = [-np.inf, -300, -250, -200, -150, -100, -50, 0, 50, 100, 150, 200, 250, 300, np.inf]
nfp_df['NFP Survey Delta Binned'] = pd.cut(nfp_df['NFP Survey Delta'], bins=bins)
nfp_grouped = nfp_df.groupby('NFP Survey Delta Binned', observed=True)
nfp_grouped_10yr = nfp_grouped[['10YR bps Delta', '2YR bps Delta']]
grouped_agg_nfp_df = pd.DataFrame(nfp_grouped_10yr.agg(['mean', 'median', 'count']).round(2))
```

```python
# Creating two separate data frames: 10YR and 2YR

grouped_agg_10yr = grouped_agg_nfp_df['10YR bps Delta']
grouped_agg2yr = grouped_agg_nfp_df['2YR bps Delta']
grouped_agg2yr
```

```python
# Graphing Grouped Nonfarm Payrolls Data for 2YR Yield Change (bps)

plt.figure(figsize=(14, 8))
ax = sns.barplot(data=grouped_agg2yr, x=grouped_agg2yr.index, y='mean')

for bar in ax.patches:
    bar.set_edgecolor('black')
    bar.set_facecolor((0.4, 0.6, 1, 0.3))

plt.title("Market Reaction to Deviation from Survey Expectation (Dec 1999 - Sept 2024)", fontsize=20, fontweight='bold')
plt.xlabel("NFP Survey Delta Binned", fontsize=12, fontweight='bold')
plt.ylabel('Average 2YR Yield Change (bps)', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')

plt.tight_layout()
# plt.savefig('/Users/henrycosentino/Desktop/2YR Average BPS Change NFP Graph.png', dpi=300)
plt.show()
```
